# 5 Perform meaningful visualizations using the dataset. Bring atleast 3 visualizations

```python
plt.title('Classification Count')
sns.countplot(file['Classification'])
```

```python
plt.figure(figsize=(8,8))
plt.title('Comparison between classification based on SL and SW',fontsize=16)
sns.scatterplot(file['SL'],file['SW'],hue=file['Classification'],s=50)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.title('Comparison between classification based on PL and PW',fontsize=16)
sns.scatterplot(file['PL'],file['PW'],hue=file['Classification'],s=50)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Classification',y='PL',data=file)
plt.subplot(2,2,2)
sns.violinplot(x='Classification',y='PW',data=file)
plt.subplot(2,2,3)
sns.violinplot(x='Classification',y='SL',data=file)
plt.subplot(2,2,4)
sns.violinplot(x='Classification',y='SW',data=file)
```

```python
plt.figure(figsize=(7,4))
sns.heatmap(file.corr(),annot=True)
```
